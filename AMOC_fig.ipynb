```python
plt.figure(figsize=(12,4),dpi=200)
plt.subplot(121)
(osf_arise_ukesm.msftyz/1026/1e6).isel(basin=0).groupby('time.year').mean('time').max('rlat').max('lev').plot(lw=3,c='b',label='ARISE-SAI-1.5')
(osf_ssp_ukesm.msftyz/1026/1e6).isel(basin=0).sel(time=slice('2015','2070')).groupby('time.year').mean('time').max('rlat').max('lev').plot(lw=3,c='r',label='SSP2-4.5')
plt.legend()
plt.ylim([12,24])
plt.xlim([2015,2070])
plt.ylabel('Sv')
plt.xlabel('Year')
plt.title('(a) UKESM1 Atlantic MOC')
plt.grid()

plt.subplot(122)
for sim in amoc_arise_cesm_arr:
    plt.plot(np.arange(2035,2070),sim[:35],lw=1,c='b',alpha=0.2)
for sim in amoc_ssp_cesm_arr:
    plt.plot(np.arange(2015,2070),sim[:55],lw=1,c='r',alpha=0.2)
plt.plot(np.arange(2035,2070),amoc_arise_cesm_ave,lw=3,c='b',label='ARISE-SAI-1.5')
plt.plot(np.arange(2015,2070),amoc_ssp_cesm_ave,lw=3,c='r',label='SSP2-4.5')
plt.legend()
plt.ylim([12,24])
plt.xlim([2015,2070])
plt.ylabel('Sv')
plt.xlabel('Year')
plt.title('(b) CESM2 Atlantic MOC')
plt.grid()
plt.show()
```
